```python
import pandas as pd

# Load the dataset
df = pd.read_csv('EVStats_AYUSH - Copy.csv')

# Check for missing values
print(df.isnull().sum())

# Rename columns for clarity
df.columns = ['SNO.','State', 'Two_Wheelers_L1_L2', 'Two_Wheelers_L2', 'Two_Wheelers_L2_250W', 'Three_Wheelers_L5_Slow', 'Three_Wheelers_L5', 'Passenger_Cars_M1', 'Buses', 'Total_In_State']
import matplotlib.pyplot as plt

# Assuming you have loaded the dataset into a DataFrame called 'df' and renamed the columns
# df.columns = ['State', 'Two_Wheelers_L1_L2', ...]

# Create a bar graph for each column
for column in df.columns[2:]:  # Exclude the 'State' column
    plt.figure(figsize=(12, 6))
    df_sorted = df.sort_values(by='SNO.', ascending=True)
    plt.bar(df_sorted['State'], df_sorted[column])
    plt.xlabel('State')
    plt.ylabel(column)
    plt.title(f'{column} by State')
    plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
    plt.tight_layout()
    #plt.savefig(f'{column}_bar_plot.png', bbox_inches='tight')
    # Show the plot
    plt.show()
    
import matplotlib.pyplot as plt

columns_to_exclude = ['SNO.','State', 'Total_In_State']
total_by_column = df.drop(columns=columns_to_exclude).sum()

# Create the pie chart
plt.figure(figsize=(8, 8))
plt.pie(total_by_column, labels=total_by_column.index, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.title('Total Numbers by Category')

# Show the pie chart
plt.show()
```
